### Problem
- Bob has a playlist of songs, each song has a singer associated with it (denoted by an integer)
- Favourite singer of Bob is the one whose songs are the most on the playlist
- Count the number of Favourite Singers of Bob

**Input Format** 
- The first line contains an integer, denoting the number of songs in Bob's playlist.
- The following input contains integers, integer denoting the singer of the song.

**Output Format**
- Output a single integer, the number of favourite singers of Bob

**Note: Use 64 bit data type**

```python
# Read number of songs
n = int(input())

# If n is 0, there are no singers to process
if n > 0:
    singers = list(map(int, input().split()))

    freq = {}  # Dictionary to store frequency

    # Count frequency of each singer
    for singer in singers:
        if singer in freq:
            freq[singer] += 1
        else:
            freq[singer] = 1

    # Find maximum frequency
    max_freq = 0
    for count in freq.values():
        if count > max_freq:
            max_freq = count

    # Count how many singers have this max frequency
    favourite_count = 0
    for count in freq.values():
        if count == max_freq:
            favourite_count += 1

    print(favourite_count)
else:
    print(0)
```

### Problem
- You are given a table with n rows m and columns. Each cell is colored with white or black. Considering the shapes created by black cells, what is the maximum border of these shapes? Border of a shape means the maximum number of consecutive black cells in any row or column without any white cell in between.

- A shape is a set of connected cells. Two cells are connected if they share an edge. Note that no shape has a hole in it.

**Input format**
- The first line contains denoting the number of test cases.
- The first line of each test case contains integers n, m denoting the number of rows and columns of the matrix. Here, '#' represents a black cell and '.' represents a white cell. 
- Each of the next n lines contains m integers.

```python
t = int(input())  # number of test cases

for _ in range(t):
    n, m = map(int, input().split())  # rows and columns
    grid = []

    for _ in range(n):
        grid.append(input().strip())

    max_border = 0

    # Check rows
    for i in range(n):
        count = 0
        for j in range(m):
            if grid[i][j] == '#':
                count += 1
                if count > max_border:
                    max_border = count
            else:
                count = 0

    # Check columns
    for j in range(m):
        count = 0
        for i in range(n):
            if grid[i][j] == '#':
                count += 1
                if count > max_border:
                    max_border = count
            else:
                count = 0

    print(max_border)
```

```python
n = int(input())

if n == 0:
    print(0)
else:
    a = list(map(int, input().split()))
    b = list(map(int, input().split()))

    min_a = min(a)  # All a[i] must be reduced to this

    total_steps = 0
    possible = True

    for i in range(n):
        diff = a[i] - min_a
        if diff < 0:
            possible = False
            break
        if b[i] == 0:
            if diff != 0:
                possible = False
                break
        elif diff % b[i] != 0:
            possible = False
            break
        else:
            total_steps += diff // b[i]

    if possible:
        print(total_steps)
    else:
        print(-1)
```

    SELECT phone_number
    FROM (
        SELECT

                CAST(SUBSTRING(duration, 1, 2) AS INTEGER) * 60 + CAST(SUBSTRING(duration, 4, 2) AS INTEGER)
            ELSE
                0
            END) AS total_incoming_duration,
            SUM(CASE WHEN call_type = 'OUT' THEN

                CAST(SUBSTRING(duration, 1, 2) AS INTEGER) * 60 + CAST(SUBSTRING(duration, 4, 2) AS INTEGER)
            ELSE
                0
            END) AS total_outgoing_duration
        FROM call_details
        WHERE call_type IN ('IN', 'OUT') -- Only consider IN and OUT calls for duration
        GROUP BY phone_number
    ) AS call_summary
    WHERE total_incoming_duration > total_outgoing_duration
    ORDER BY total_incoming_duration DESC;

    from typing import List

    def missingNumber(nums: List[int]) -> int:
        """
        Finds the one missing number in an array containing n distinct numbers
        taken from the range [0, n].
        """
        n = len(nums)
        # Calculate the expected sum of numbers from 0 to n
        expected_sum = n * (n + 1) // 2
        # Calculate the actual sum of numbers in the given list
        actual_sum = sum(nums)
        # The missing number is the difference between the expected sum and the actual sum
        return expected_sum - actual_sum

    def fibonacci(n: int) -> int:
        """
        Generates the n-th Fibonacci number.
        """
        if n <= 1:
            return n
        else:
            a, b = 0, 1
            for _ in range(2, n + 1):
                a, b = b, a + b
            return b

    # Example usage:
    # n = 5
    # print(fibonacci(n)) # Output: 5

    # n = 0
    # print(fibonacci(n)) # Output: 0

    def runningAverage() -> Generator[float, int, None]:
        """
        Coroutine to compute the running average of a stream of numbers.
        """
        total = 0
        count = 0
        average = 0.0
        while True:
            try:
                # Yield the current average and wait for the next number
                number = yield average
                total += number
                count += 1
                average = total / count

                pass
